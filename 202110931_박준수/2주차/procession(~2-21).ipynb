```python
import numpy as np
A=np.array([[1,2,3,4],[5,6,7,8]])
print(A[0])
print(A[0,:])
print(A[1])
print(A[1,:])
print(A[:,0])
print(A[:,1])
print(A[:,2])
print(A[:,3])

print('\n\n')

A=np.array([[1,2],[3,4]])
B=np.array([[4,1],[5,2]])

print(A+B)
print(np.add(A,B))
print(A-B)
print(np.subtract(A,B))

print('\n\n')

A=np.arange(9).reshape(3,3)
print(A)
c=10
print(np.multiply(c,A))
print(c*A)

print('\n\n')

A=np.array([[1,2,3],[4,5,6]])
B=np.array([[4,1],[5,2],[1,2]])
print(A)
print(B)
print(A.dot(B))
print(np.dot(A,B))

print('\n\n')

A=np.array([[1,2],[3,4]])
B=np.array([[4,1],[5,2]])
print(A*B)
print(np.multiply(A,B))

print('\n\n')

X=np.array([[1,2],[3,4],[5,6]])
print(X)
print(X.T)
print(np.transpose(X))

print('\n\n')

I=np.eye(3)
A=np.arange(9).reshape(3,3)
print(A)

print((np.dot(A,I)==A).all())
print((np.dot(I,A)==A).all())

print('\n\n')

A=np.array([[1,2],[3,4]])
invA=np.linalg.inv(A)
print(invA)

print(np.dot(A,invA))
print(np.round(np.dot(A,invA)))
print(np.round(np.dot(invA,A)))
```
