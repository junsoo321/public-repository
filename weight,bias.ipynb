```python
import numpy as np
import matplotlib.pyplot as plt
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
x=np.arange(-10,10,0.1)
I=np.arange(0.5,3,0.5)
J=np.arange(-2,2,0.5)
for i in I:
  for j in J:
    f=1/(1+np.exp(-x*i+j))
    ax.plot(x,f)
ax.set_xticks(range(-10,10))
ax.set_yticks(range(0,2))
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
plt.show
```
